```python
class Demo:
    a = 100
    def __init__(self,a):
        self.a = 500 
    def printData(self):
        print(self.a)
    @classmethod
    def classMethod(cls):
        print(cls.a)

    @staticmethod
    def staticMethod(b):
        print(b)

d1 = Demo(10)
d1.printData()
Demo.classMethod()
d1.staticMethod(10)
```

```python
class Time:
    def __init__(self,*t):
        if(len(t)==2):
            self.h = t[0]
            self.m = t[1]
        else:
            self.h = t[0].h
            self.m = t[0].m,

    def printTime(self):
        print(self.h,self.m)

t1 = Time(5,6)
t2 = Time(t1)
t2.printTime()
```
