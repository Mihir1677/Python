### 04) WAP to find the longest word in a file named abc.txt

```python
file1 = 'file1.txt'
file = open(file1,'r')
data = file.read().split();
max = data[0];
for i in data:
    if(len(max)<len(i)):
        max = i
print(max)
```

### 01) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
file1 = 'primenumbers.txt'
file = open(file1,'w')
isPrime = False;
for i in range(2,101):
    for j in range(1,i):
        isPrime = False
        if(i%j==0 and j!=1):
            isPrime = True;
    if(isPrime):
        file.write(str(i));
```
